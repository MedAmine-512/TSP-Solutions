## Brute-Force Resolution

```python

def perm(L):
    if len(L) == 1 or len(L) == 0:
        return [L]
    out =[] 
    for i in range(len(L)) :
        R = L[:i] + L[i+1:]
        for j in perm(R):
            out.append(j+[L[i]])
    return out

        
L = [0, 1, 2, 3]
perm(L)

```

```python
def loops(n, start):
    L = perm([i for i in range(n)])
    return [i + [i[0]] for i in L if i[0] == start]
    

loops(10,3)
```

```python
def distance(LB, T):
    out = []
    for i in LB :
        dist = 0
        for j in range(len(i)-1):
            dist += T[i[j]][i[j+1]]
        out.append(dist)

    return out 


LB = loops(4, 2)

T = [
    [0, 5, 4, 9],
    [3, 0, 4, 7],
    [5, 1, 0, 2],
    [6, 2, 8, 0]
]

print(distance(LB, T))
```

```python
def best(LB,T):
    y = distance(LB,T)
    x = min(y)
    return (x, LB[y.index(x)])
best(LB,T)
```
